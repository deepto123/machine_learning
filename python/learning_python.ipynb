#PYTHON LOOPS

```python
#simple while loop

i=1

while i<=5:
  print(i)
  i+=1  #or i=i+1

print('Loop Terminated')
```

```python
#finding sum of even numbers between 1 and 100

i=1;
sum=0;

while i<=100:
  if (i%2)==0:
    sum+=i
  i+=1

print(sum)
```

```python
%%time
#calculate factorial of a given number

val=int(input('Enter the number to find its factorial:'))

if(val==1):
  print(0)
  raise SystemExit("Stop right there!")

prd=1

for i in range(val):
  prd*=(i+1)

print(prd)
```

```python
  #factorial using recursion

def fact(n):
   if n==0:
    return 1

   return n*fact(n-1)
    
x=fact(20)
print(x)
```

```python
#Arbitrary Arguments, *args
#If you do not know how many arguments that will be passed into your function, add a * before the parameter name in the function definition.


def my_function(*kids):
  print("The youngest child is " + kids[2])

my_function("Emil", "Tobias", "Linus")

#this is bcoz here the parameter list could have been 1,2,4,5,6 ans so on anything
```

```python
#Keyword Arguments
#You can also send arguments with the key = value syntax.

#This way the order of the arguments does not matter.

def op(a,b,c):
  print(a+b-c)

op(1,4,3)
op(c=1,b=3,a=4)
```

```python
#Arbitrary Keyword Arguments, **kwargs
#If you do not know how many keyword arguments that will be passed into your function, add two asterisk: ** before the parameter name in the function definition.
#This way the function will receive a dictionary of arguments, and can access the items accordingly:

def op(**val):
  print(val['a']+val['b']-val['c'])

#op(1,4,3) errpr
op(c=1,b=3,a=4)
```
